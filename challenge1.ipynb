```python
import os
import shutil
```

```python
path = 'G:'
list_ = os.listdir(path)
```

```python
list_
```

```python
for file_ in list_:
    name, ext = os.path.splitext(file_)
    print(name,ext)
    ext = ext[1:]
```

```python
for file_ in list_:
    
    name, ext = os.path.splitext(file_)
    ext = ext[1:]
    if ext == '':
        continue
    if os.path.exists(path+'/'+ext):
        shutil.move(path+'/'+file_, path+'/'+ext+'/'+file_)
    else:
        os.makedirs(path+'/'+ext)
        shutil.move(path+'/'+file_, path+'/'+ext+'/'+file_)
```
